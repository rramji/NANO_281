#### Q1.1

#### Q1.4

```python
joules_per_ev = 1.60218e-19
atoms_per_mol = 6.022e23
df['e_form_j/mol'] = df['formation_energy_per_atom'] * joules_per_ev * atoms_per_mol
df.head()
```

#### Q1.5
Let us assume that materials with energy above hull of >0.03 eV/atom are unstable and are potentially stable otherwise. 
Furthermore, band gaps of 0, (0, 1], [1, +inf) are indicative of metallic, small band gap, large band gap for the materials.
- Create a table of the number of ABO3 crystals in each joint category, e.g., (unstable, metallic), (unstable, small band gap), etc.

```python
df.loc[df['band_gap'] == 0, 'bg_class'] = 'metallic'
df.loc[df['band_gap'] > 0, 'bg_class'] = 'small band gap'
df.loc[df['band_gap'] > 1, 'bg_class'] = 'large band gap'
df.loc[df['e_above_hull'] > 0.03, 'stability'] = 'unstable'
df.loc[df['e_above_hull'] <= 0.03, 'stability'] = 'potentially stable'
df.groupby(['stability', 'bg_class']).size()
```

#### Q3.3
- Perform a hypothesis test at the 95% level to determine if there is a significant difference between the formation energies reported in the Materials Project and the OQMD dataset (hint: check out the scipy.stats.ttest_ind method). Discuss your findings, including providing any possible explanations for any discrepancy between the two datasets.


```python
from scipy import stats
#test if datasets have identical variances
t, p = stats.bartlett(oqmd_eng, mp_eng)
if p > 0.05:
    print("Fail to reject null hyp, variances should be treated as equal\n")
    t_stat, p_val = stats.ttest_ind(oqmd_eng, mp_eng)
else:
    print("Reject null hyp, variances are different\n")
    t_stat, p_val = stats.ttest_ind(oqmd_eng, mp_eng, equal_var=False)

if p_val > 0.05:
    print("Fail to reject null hyp, no significant difference \
    \nbetween formation energies in these datasets")
else:
    print("Reject the null hyp, there *is* a statistically significant difference\
    \nbetween formation energies in these datasets of", round(t_stat,3))
```
